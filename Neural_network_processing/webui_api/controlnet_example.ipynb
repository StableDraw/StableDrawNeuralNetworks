### Controlnet

Для работы необходимо изображение (можно сгенерировать).

Препроцессор (module) - извлекает из рисунка информацию / изменяет изображение в зависимости от типа (границы, поза и т.д.).

Модель (model) - по извлеченной информации и промту генерирует новую картинку.

img2img версия так же использует картинку для генерации.

```python
result1 = web_api.text2img(prompt="squirrel",
                    negative_prompt="ugly, out of frame",
                    seed=700,
                    cfg_scale=7,
                    )
result1.image
```

```python
result2 = web_api.text2img_controlnet(
    input_image=result1.image,
    prompt="bird",
    negative_prompt="ugly, out of frame",
    module="softedge_pidinet",
    model="control_v11p_sd15_softedge"
    )
```

```python
result2.images[0]
```

```python
result2.images[1]
```

```python
result3 = web_api.text2img(prompt="bird",
                    negative_prompt="ugly, out of frame",
                    seed=500,
                    cfg_scale=7,
                    )
result3.image
```

```python
image1 = PIL.Image.open('colors1.png')
```

```python
image1
```

```python
result4 = web_api.img2img_controlnet(
    images=[image1],
    input_image=result3.image,
    module="softedge_pidinet",
    model="control_v11p_sd15_softedge",
    pixel_perfect = True,
    weight=1
    )
```

```python
result4.images[0]
```
