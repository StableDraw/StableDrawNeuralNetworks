```python
from typing import List
import io
from PIL import Image
import webui
```

```python
caption = "squirrel"
params = {
    "negative_prompt": "ugly, out of frame",
    "seed": 700,
    "cfg_scale": 7,
    }
binary_result1 = webui.webui_text2img(caption, params)
result1 = Image.open(io.BytesIO(binary_result1))
result1
```

```python
input_data = [result1]
input_binary_data = []
for img in input_data:
        img_byte_arr = io.BytesIO()
        img.save(img_byte_arr, format='PNG')
        input_binary_data.append(img_byte_arr.getvalue())
caption = "bird"
params = {
    "negative_prompt": "ugly, out of frame",
    "seed": 700,
    "cfg_scale": 7,
    }
binary_result2 = webui.webui_img2img(input_binary_data, caption, params)
result2 = Image.open(io.BytesIO(binary_result2))
result2
```

```python
img_byte_arr = io.BytesIO()
result1.save(img_byte_arr, format='PNG')
input_binary_data = img_byte_arr.getvalue()
caption = "bird"
params = {
    "negative_prompt": "ugly, out of frame",
    "module": "softedge_pidinet",
    "model": "control_v11p_sd15_softedge",
    "pixel_perfect": True,
    }
binary_result3 = webui.webui_text2img_controlnet(input_binary_data, caption, params)
result3 = Image.open(io.BytesIO(binary_result3[0]))
result3
```
