```python
caption = "squirrel"
params = {
    "negative_prompt": "ugly, out of frame",
    "seed": 700,
    "cfg_scale": 7,
    "version": "SDXL-base-1.0",
    "low_vram_mode": False
    }
binary_result1 = INeural.text_to_image(caption, params)
result1 = Image.open(io.BytesIO(binary_result1))
result1
```

```python
input_data = [result1]
input_binary_data = []
for img in input_data:
        img_byte_arr = io.BytesIO()
        img.save(img_byte_arr, format='PNG')
        input_binary_data.append(img_byte_arr.getvalue())
caption = "bird"
params = {
    "negative_prompt": "ugly, out of frame",
    "seed": 700,
    "cfg_scale": 7,
    }
binary_result2 = INeural.webui_img2img(input_binary_data[0], caption, params)
result2 = Image.open(io.BytesIO(binary_result2))
result2
```

```python
params = {
    "upscaling_resize" : 1.5,
    "upscaler_1" : "LDSR"
}
binary_result3 = INeural.webui_upscaler(binary_result2, params)
result3 = Image.open(io.BytesIO(binary_result3))
print(result3.size)
result3
```
